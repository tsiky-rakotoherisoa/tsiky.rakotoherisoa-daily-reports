### Exercise 1:

### Exercise 16:

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---


````python
x = np.linspace(0, 10, 400)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: y = x
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')


# Second subplot: y = x
axs[0, 1].plot(x, y2, color='green')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

# Third subplot: y = x
axs[1, 0].plot(x, y3, color='red')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Fourth subplot: y = x
axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')


plt.show()
````

Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.

```python
# Exercise 21
```

---


````python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

# Plot
plt.figure(figsize=(6, 6))
contour = plt.contour(X, Y, Z, cmap='coolwarm')
plt.clabel(contour, inline=True, fontsize=8)

plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(contour)
plt.show()
````


### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.

```python
# Exercise 26
```

---


````python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='inferno', edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()

````
